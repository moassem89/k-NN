```python
import numpy as np
```

```python
r = np.random.rand(1000000)*0.25  # sampling random array of values between 0, 0.25 to be used later
```

```python
for d  in range(2,100):

    basis = np.eye(d) # array of the standard basis vectors for R^d
    
    appendix = r[:d-1] # a random array of appropriate length

    t = np.concatenate(([1] , appendix)) # an arbitrary point closer to u than v of dimension d
    
    u = basis[0]
    v = basis[d-1]

    print(np.linalg.norm( t - v) /  np.linalg.norm( t - u ))
```
